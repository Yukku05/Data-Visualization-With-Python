```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# load the dataset
df = pd.read_csv("Iris.csv")
```

```python
# Remove the 'Id' column
df.drop(columns=['Id'], inplace=True)
```

# 1. Box Plot for each feature
plt.figure(figsize=(12, 6))
df.drop(columns=['Species']).boxplot()
plt.title("Box Plot of Iris Features")
plt.ylabel("cm")
plt.grid(True)
plt.show()

```python
# 2. Scatter Plot (Sepal Length vs Sepal Width)
species = df['Species'].unique()
colors = ['red', 'green', 'blue']
```

```python
plt.figure(figsize=(8, 5))
for sp, color in zip(species, colors):
    subset = df[df['Species'] == sp]
    plt.scatter(subset['SepalLengthCm'], subset['SepalWidthCm'], label=sp, color=color)

plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 3. Histogram for Petal Length
plt.figure(figsize=(8, 5))
plt.hist(df['PetalLengthCm'], bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# 4. Pie Chart of Species Count
species_count = df['Species'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(species_count, labels=species_count.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Iris Species")
plt.show()
```
